**3.1. Agrupando Colunas**

**3.1.1. Relação entre QUANTIDADE DE CARTOES E SEUS TIPOS COM OS CLIENTES**

**a) Qunatidade de cartões separados por tipos**

query: select tipo_cartao, count(tipo_cartao) as QTD_CARTOES from credito group by tipo_cartao











**b) Quantos cartões temos separados por tipo, quantidade de cartões e sexo?**

query: select tipo_cartao, count(tipo_cartao) as QTD_CARTOES, sexo from credito group by tipo_cartao, sexo;







**c) Quantidade de clientes separados por sexo**

query: SELECT count(*) as Contagem_de_clientes, sexo from credito group by sexo






> Na primeira query **a)** nós vemos, claramente, que o cartão mais usado em uma porporção muito maioir é o **blue**.

> No caso dos cartões divididos por sexo, **a proporção é obedecida**, porém, as mulheres tem um aumento leve da variação com relação aos homens.

**3.1.2. Relação entre SALÁRIO ANUAL/SEXO/TIPO DE CARTÃO**




query: SELECT count(*) as Contagem_clientes, salario_anual, sexo from credito group by salario_anual,sexo











> Aqui as discrepâncias aparecem de forma gritante!
Temos muitos clientes do sexo feminino que não possuem dados de salário anual, então essa amostra de dados **não nos trará uma análise tão confiável**, mas sim possíveis indicadores, porém com os dados presentes é possível perceber que a disseminação dos clientes masculinos sobre as faixas é muito maior que o feminino.


> Também percebemos que somente os clientes **MASCULINOS**  estão acima da faixa dos 60k anuais. Isso pode indicar a discrfepância de valores pagos como salários para homens e mulheres na sociedade.

> Contudo devemos ter cuidado, pois a **quantidade de dados faltantes** masculinos é insignificante,mas a quantidade de dados faltantes femininos é muito significativa.

**3.1.3. Relação entre limite de crédito/escolaridade/tipo de cartão/sexo**





> Temos a **tendência de pensar que pessoas com um maior nível de escolaridade possuem maiores salários** e,por isso, maiores limites de crédito, mas nessa amostra, **isso não se traduz em fatos**.

> A **escolaridade não é um fator importante** na classificação dos limites de crédito.

> Porém podemos observar que **os maiores limites estão vinculados ao sexo masculino**.

> **Isso já era esperado**, visto que a faixa de salário anual eram maiores para os indivíduos de sexo masculino.

**3.1.4. Relação entre Media da quantidade de produtos/Media do valor das transações
/Media da inatividade nos ultimos 12m/Iteracoes nos 12 últimos meses/sexo/salario anual**

query: select avg(qtd_produtos) as Media_qts_produtos, avg(valor_transacoes_12m) as Media_valor_transacoes, avg(meses_inativo_12m) as Media_inatividade_nos_ultimos_12m, avg(iteracoes_12m) as Iteracoes_12m, sexo, salario_anual from credito where salario_anual != 'na' group by sexo, salario_anual order by avg(valor_transacoes_12m) desc





> Essa tabela traz informações extremamente interessantes

> Os **valores das transções são maiores para os que possuem menores rendas** anuais e diminuem a partir do aumento da renda anual

> O valor médio das **quantidades de produtos** permanecem próximas. **Não é um fator relevante**


> A quantidade média de **inatividade** é próxima. **Não é um fator relevante**


> A média de **iterações** é próxima. **Não é um fator relevante**



**3.1.5. Relação entre dependentes e limite de crédito**


query: select avg(limite_credito) as Media_limite_credito, avg(dependentes) as dependentes from credito group by dependentes;







> Essa é uma informação muito imporatante. **Quanto maior o número de dependentes, maior será o limite de credito**

**3.1.6. Relação entre limite de credito/estado civil**


query: select avg(limite_credito) as Media_limite_credito, estado_civil from credito group by estado_civil;




> Temos muitos valores nulos, mas os **solteiros e divorciados possuem uma leve vantagem de crédito em relaçãos aos casados**

**4. Insights and Conclusion**
------



Considerando este dataset, temos:
* Apesar de algumas análises do público feminino apresentarem uma falta de dados importante, conseguimos notar que **os maiores salários são dados aos homens**
* A quantidade de limite de crédito tem uma relação direta com a renda anual, ou seja, **quanto maior a renda maior o limite de crédito **
* Em média, **quanto mais dependentes tiver o cliente maior o seu limite de crédito**. Faz sentido se considerarmos que mais dependentes, proporcionam mais gastos
* A maior parte dos clientes são masculinos, porém os clientes femininos não passam dos 60k anuais de ganhos
* A aior parte dos clientes possui renda até 40K e é composta por mais mulheres
* A quantidade de **credito não tem nada a ver com escolaridade** dos clientes
* Possuímos clientes de pouco mais de 20 anos até pocuo mais que 70. **Os maiores limites estão com uma faixa média entre os 40 a 50 anos**
* A maioria das amostras que temos dos tipos de cartão são blue. **Temos poucas informações sobre os outros tipos de cartão além do blue**
* Solteiros e divorciados possuem uma leve vantagem de crédito em relaçãos aos casados
* A quantidade média de produtos são próximas ou seja, **mulheres e homens tem basicamente os mesmos gastos**
